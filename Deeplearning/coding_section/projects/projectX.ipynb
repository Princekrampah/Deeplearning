```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow import keras
```

```python
df = pd.read_csv("../../../datasets/science_students.csv")
df = df.drop(['Unnamed: 0', 'index'], axis=1)
```

```python
df.head()
```

```python
df = df[df['Course'].isin(['PCM', 'PGM', 'CBG', 'EGM', 'PCB'])]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_div = LabelEncoder()
l_civ = LabelEncoder()
l_hist = LabelEncoder()
l_geo = LabelEncoder()
l_kisw = LabelEncoder()
l_eng = LabelEncoder()
l_phy = LabelEncoder()
l_chem = LabelEncoder()
l_bio = LabelEncoder()
l_math = LabelEncoder()
l_combination = LabelEncoder()
```

```python
df['DIV'] = l_div.fit_transform(df['DIV'])
df['CIV'] = l_civ.fit_transform(df['CIV'])
df['HIST'] = l_hist.fit_transform(df['HIST'])
df['GEO'] = l_geo.fit_transform(df['GEO'])
df['KISW'] = l_kisw.fit_transform(df['KISW'])
df['ENGL'] = l_eng.fit_transform(df['ENGL'])
df['PHY'] = l_phy.fit_transform(df['PHY'])
df['CHEM'] = l_chem.fit_transform(df['CHEM'])
df['BIO'] = l_bio.fit_transform(df['BIO'])
df['B/MATH'] = l_math.fit_transform(df['B/MATH'])
df['Course'] = l_combination.fit_transform(df['Course'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Course"], axis=1), df['Course'], train_size=0.8)
```

```python
df.head()
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(11, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(100, input_shape=(100, ), activation='sigmoid'),
    keras.layers.Dense(5 ,activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=200)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
y_predicted_labels = [np.argmax(n) for n in y_pred]
```

```python
cm = tf.math.confusion_matrix(y_test.values, y_predicted_labels)
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
plt.figure(figsize=(12, 7))
sn.heatmap(cm, annot=True)
plt.ylabel("Truth")
plt.xlabel("Predictions")
plt.show()
```
