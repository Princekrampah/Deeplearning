```python
import numpy as np
np.random.seed(0)
```

```python
class NeuralNetwork:
    def __init__(self, n_inputs, n_neurons, actual):
        self.weights = 0.1*np.random.rand(n_inputs, n_neurons)
        self.actual = actual
        self.outputs = np.zeros(n_neurons)
        
    def sigmoid(self, x):
        self.z = 1 / (1 + np.exp(x))
        return self.z
    
    def feedforward(self, inputs):
        self.inputs = inputs
        self.outputs = self.sigmoid(np.dot(inputs, self.weights))
        
    def sigmoid_derivative(self, x):
        return self.sigmoid(x)*(1 - self.sigmoid(x))
        
    def backprop(self):
        for i in range(10):
            self.cost = self.actual - self.sigmoid(self.outputs)
            print(self.cost)
```

```python
layer1 = NeuralNetwork(3, 1, [1, 3, 4])
```
