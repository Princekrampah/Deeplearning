# Customer Churn Prediction 

Predict behavior to retain customers

**The data includes the following:**

- Customers who left within the last month – the column is called Churn

- Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies

- Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly charges, and total charges

- Demographic info about customers – gender, age range, and if they have partners and dependents

- Tenure - the number of months the customer subscribed to the service.

[Source - Details on dataset](https://www.kaggle.com/blastchar/telco-customer-churn)

```python
for column in df:
    print(f"{column} : {df[column].unique()}")
```

```python
df.replace("No internet service", "No", inplace=True)
df.replace("No phone service", "No", inplace=True)
df
```

```python
categorical_cols = []

for column in df:
    if (df[column].dtype == "object"):
        if("Yes" in df[column].unique() and "No" in df[column].unique()):
            df[column].replace({"Yes" : 1, "No" : 0}, inplace=True)
        elif ("Female" in df[column].unique() and "Male" in df[column].unique()):
            df[column].replace({"Female" : 0, "Male" : 1}, inplace=True)
        else:
            categorical_cols.append(column)
        
 
```

```python
df
```

```python
df.info()
```

```python
categorical_cols
```

```python
df = pd.get_dummies(data=df, columns=categorical_cols)
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
```

```python
scaled_cols = ["tenure", "MonthlyCharges", "TotalCharges"]


df[scaled_cols] = min_max_scaler.fit_transform(df[scaled_cols])
```

```python
df
```

```python
df['MonthlyCharges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Churn'], axis="columns")
y = df["Churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

## Deep Learning Neural Network

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(units=24, input_shape=(26, ), activation="relu"),
    Dense(units=15, activation="relu"),
    Dense(units=1, activation="sigmoid")
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = [1 if x > 0.5 else 0 for x in y_pred]
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

## Model Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = tf.math.confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(12, 7))
plt.title("Confusion Matrix", fontsize=18)
sn.heatmap(cm, annot=True)
plt.xlabel("Prediction")
plt.ylabel("Truth")
plt.show()
```
